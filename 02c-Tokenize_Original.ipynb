## Tokenize

* Load datasets
* Tokenize
* Export Tensor for Test and Train sets to .pt files

```python
import torch
from transformers import (
    BertTokenizer,
)

from shared import create_tensor_dataset, BERT_MODEL, train_dataset_path, test_dataset_path, \
    tokenized_train_dataset_path, tokenized_test_dataset_path
```

```python
# Import splits
train = torch.load(train_dataset_path)
test = torch.load(test_dataset_path)
```

```python
tokenizer = BertTokenizer.from_pretrained(BERT_MODEL)
```

```python
# Tokenize
train_dataset = create_tensor_dataset(train)
test_dataset = create_tensor_dataset(test)
```

```python
# Export
torch.save(train_dataset, tokenized_train_dataset_path)
print(f"Training dataset saved to {tokenized_train_dataset_path}")
torch.save(test_dataset, tokenized_test_dataset_path)
print(f"Test dataset saved to {tokenized_test_dataset_path}")
```
